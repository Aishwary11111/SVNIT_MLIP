```python
from ultralytics import YOLO

# Function to train the model
def train_model(total_epochs, image_size, model_path="detection_model.pt", config_yaml='config.yaml'):
    detection_model = YOLO(model_path) 
    detection_model.train(data=config_yaml, epochs=total_epochs, imgsz=image_size)
    return detection_model

# Uncomment the following line and specify the required parameters to train the model.    
# train_model(total_epochs=100,image_size=224,model_path='detection_model.pt', config_yaml='config.yaml')
```
